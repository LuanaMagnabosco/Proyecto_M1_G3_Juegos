## JUEGO 1: PIEDRA PAPEL TIJERA

```python
#ingresar nombres de jugadores
nombre1 = input("Ingresa tu nombre: ")
nombre2 = input("Ingresa tu nombre: ")

#variable opciones para el while que verifica si se escribiò correctamente piedra-papel-tijera
opciones = ["piedra", "papel", "tijera"]

#variables para guardar los puntajes de cada jugador
puntaje_jugador1 = 0
puntaje_jugador2 = 0


#mientras ningun jugador sume 3 puntos, entra al bucle, cuando alguno cumpla 3 victorias, se termina
while (puntaje_jugador1 < 3) and (puntaje_jugador2 < 3):
    #eleccion de piedra, papel o tijera
    jugador1 = input("Elige entre piedra, papel o tijera").lower()
    jugador2 = input("Elige entre piedra, papel o tijera").lower()
    
    #si ingresa una opción inválida, se le vuelve a pedir que ingrese piedra, papel o tijera
    while jugador1 not in opciones or jugador2 not in opciones:
        if jugador1 not in opciones: 
            jugador1 = input(f"{nombre1},opcion inválida, vuelve a intentarlo elige entre piedra, papel o tijera").lower()
        elif jugador2 not in opciones:
            jugador2 = input(f"{nombre2}, opcion inválida, vuelve a intentarlo elige entre piedra, papel o tijera").lower()
        
    #comparaciones
    if jugador1 == jugador2:
        print("Empate")
    elif jugador1 == "piedra" and jugador2 == "papel":
     print(f"{nombre1} eligió piedra")
     print(f"{nombre2} eligió papel")
     print(f"Partida ganada por {nombre2}")
     #suma puntaje a quien corresponda
     puntaje_jugador2 += 1
    elif jugador2 == "piedra" and jugador1 == "papel":
        print(f"{nombre1} eligió papel")
        print(f"{nombre2} eligió piedra")
        print(f"Partida ganada por {nombre1}")
        #suma puntaje a quien corresponda
        puntaje_jugador1 += 1
    elif jugador1 == "piedra" and jugador2 == "tijera":
        print(f"{nombre1} eligió piedra")
        print(f"{nombre2} eligió tijera")
        print(f"Partida gnada por {nombre1}")
        #suma puntaje a quien corresponda
        puntaje_jugador1 += 1
    elif jugador2 == "piedra" and jugador1 == "tijera":
        print(f"{nombre1} eligió tijera")
        print(f"{nombre2} eligió piedra")
        print(f"partida ganada por {nombre2}")
        #suma puntaje a quien corresponda
        puntaje_jugador2 += 1
    elif jugador1 == "papel" and jugador2 == "tijera":
        print(f"{nombre1} eligió papel")
        print(f"{nombre2} eligió tijera")
        print(f"Partida ganada por {nombre2}")
        #suma puntaje a quien corresponda
        puntaje_jugador2 += 1
    elif jugador2 == "papel" and jugador1 == "tijera":
        print(f"{nombre1} eligió tijera")
        print(f"{nombre2} eligió papel")
        print(f"partida ganada por {nombre1}")
        #suma puntaje a quien corresponda
        puntaje_jugador1 += 1
        
#mensaje de fin del juego
     
print("El juego acabó! El puntaje es el siguiente: \n")
print(f"{nombre1}: {puntaje_jugador1}")
print(f"{nombre2}: {puntaje_jugador2}")


#comparación de puntajes para saber el ganador
if puntaje_jugador1 > puntaje_jugador2:
    print(f"El ganador es {nombre1}")
else:
    print(f"El ganador es {nombre2}")
```

## JUEGO 2: PREGUNTAS Y RESPUESTAS GEOGRAFÍA DE ESPAÑA

```python
preguntas = {
            "El pico más alto de España peninsular es el Mulhacén?" : "verdadero",
            "El Pico Viejo es el segundo pico más alto de España?": "verdadero", 
            "Tiene 2 ciudades autónomas ?": "verdadero",
            "Es el rio Tajo el más largo de los ríos de España ?" : "verdadero", 
            "Es el río Ebro el más caudoloso?": "verdadero",
            "El Pico del Teide es un volcán?": "verdadero",
            "El Río Guadalquivir se encuentra en Andalucía?": "verdadero",
            "Oviedo se encuentra en Asturias?": "verdadero",
            "Tiene 45 provincias?" : "falso", 
            "Tiene 15 comunidades autónomas ?": "falso", 
            "La ciudad de las Artes y las Ciencias está en Barcelona?": "falso",
            "La Sagrada Familia se encuentra en Madrid?": "falso",
            "El Palacio de Cristal se encuentra en Valencia?": "falso",
            "Los tres picos más altos de España se encuentran fuera de la península?" : "falso", 
            "Lugo está en Asturias?": "falso",
            "Cantabria está en el País Vasco?": "falso",
}
```

```python
import random 

#el parámetro dict por si se quiere jugar con un diccionario diferente
def seleccion(dict):
    #pasar a lista el dict para poder guardar por un lado la pregunta y por otro lado la respuesta
    lista = list(dict.items())
    #elección aleatoria de la pregunta    
    pregunta = random.choice(lista)
    respuesta = pregunta[1]
    #como trabajamos con lista, que se guarde la pregunta por un lado y la respuesta por otro
    return pregunta[0], respuesta 
```

```python
nombre = input("Ingresa tu nombre")
correcta = 0
incorrecta = 0
partida = 0
opcion = ["verdadero", "falso"]

#mientras las respuestas correctas sean menos de 5 y las incorrectas menos de 3
while (correcta < 5) and (incorrecta < 3):
    pregunta, respuesta = seleccion(preguntas)
    #conteo de partidas
    partida += 1
    print("-------------------------")
    #imprime la cantidad de partidas escondida en números de preguntas
    print(f"{nombre} pregunta número: {partida}")
    print(pregunta)
    jugador = (input("Verdadero o falso?: ").lower())
    #verificar que esté bien escrito, comparando con la lista opciones
    if jugador not in opcion:
        escrito_bien = input(f"{nombre},Vuelve a intentarlo. Ingresa verdadero o falso").lower()
        if escrito_bien == respuesta:
            print(f"Tu respuesta: {escrito_bien}.\n Respuesta correcta")
            correcta += 1
        else:
            print(f"Tu respuesta: {escrito_bien}. \n Respuesta incorrecta")
            incorrecta += 1       
    #si el jugador acierta la respuesta
    elif jugador == respuesta:
        print(f"Tu respuesta: {jugador}.\n Respuesta correcta")
        correcta += 1
    #si el jugador no acierta
    else:
        print(f"Tu respuesta: {jugador}. \n Respuesta incorrecta")
        incorrecta += 1


print("------------------")        
print("El juego terminó!")
if correcta == 5:
    print(f"Enhorabuena {nombre}! Has ganado el juego.")
else:
    print(f"{nombre} has perdido. Mejor suerte la próxima vez.")
print(f"Tu resultado es el siguiente: \n Preguntas correctas: {correcta}. \n Preguntas incorrectas: {incorrecta}.")
```

## JUEGO 3: AHORCADO

```python
# Importamos desde random el método choice para elegir una palabra aleatoria dentro da la lista de palabras.
from random import choice

#Mensaje de bienvenida al juego.
print("\n")
print("¡Bienvenido al juego del Ahorcado!\n")
print("Intenta adivinar la palabra oculta antes de que se acaben tus vidas.\n")
print("Tienes 6 vidas.\n")
print("¡Buena suerte!\n")

#Definimos una variable que sea una lista de palabras para utilizar el método choice.
palabras = ['adalab', 'sql', 'analytics', 'data']

#Creación de las variables globales, listas para almacenar las letras correctas e incorrectas y los contadores de intentos y vidas.
letras_correctas = []
letras_incorrectas = []
intentos = 6
aciertos = 0
#Establecer la variable como False para poder ejecutarse el bucle hasta que se cambie dicha condición.
juego_terminado = False

#Función para elegir una palabra aleatoria y organizar la separación de las letras. 
def elegir_palabra(lista_palabras):
    palabra_elegida = choice(lista_palabras)
    letras_de_palabra = len(set(palabra_elegida))

    return palabra_elegida, letras_de_palabra

#Función que contiene el abecedario y el input de inicio de juego, establecemos la variable es_valida como False para que entre en el bucle y cuando cambie dicha condición el bucle se interrumpa.
def pedir_letra():
    letra_elegida = ''
    es_valida = False
    abecedario = "abcdefghijklmnñopqrstuvwyz"

    while not es_valida:
        letra_elegida = input("Elige una letra: ").lower()
        if letra_elegida in abecedario and len(letra_elegida) == 1:
            es_valida = True
        #Else se imprime cuando el jugador introduzca más de una letra o un caracter diferente de una letra.
        else:
            print("Solo puedes introducir letras.")

    return letra_elegida

#Función muestra el estado actual del tablero del juego, enseñando las letras correctas adivinadas y los espacios en blanco para las letras que faltan.
def mostrar_nuevo_tablero(palabra_elegida):

    lista_oculta = []
    #Bucle for para que itere la lista y, dependiendo si la letra está o no en la lista oculta se añade y devuelve o la letra adivinada o un guión.
    for l in palabra_elegida:
        if l in letras_correctas:
            lista_oculta.append(l)
        else:
            lista_oculta.append('-')

    print(' '.join(lista_oculta))
    #Ponemos el dibujo del muñeco dentro de la funcion tablero y se imprime segun el contador de intentos.
    if intentos == 6:
        print("  ___")
        print(" |       |")
        print(" |")
        print(" |")
        print(" |")
        print("_|_")
        
    elif intentos == 5:
        print("  ___")
        print(" |       |")
        print(" |       O")
        print(" |")
        print(" |")
        print("_|_")
    elif intentos == 4:
        print("  ___")
        print(" |       |")
        print(" |       O")
        print(" |       |")
        print(" |")
        print("_|_")
    elif intentos == 3:
        print("  ___")
        print(" |       |")
        print(" |       O")
        print(" |      /|")
        print(" |")
        print("_|_")
    elif intentos == 2:
        print("  ___")
        print(" |       |")
        print(" |       O")
        print(" |      /|\\")
        print(" |")
        print("_|_")
    elif intentos == 1:
        print("  ___")
        print(" |       |")
        print(" |       O")
        print(" |      /|\\")
        print(" |      /")
        print("_|_")
    

#Función para comprobar si la letra elegida por el jugador está presente en la palabra oculta, y actualiza el estado del juego en consecuencia. 
def comprobar_letra(letra_elegida, palabra_oculta, vidas, aciertos, letras_de_palabra):
    #Creamos una variable = False para indicar que el juego no ha terminado.
    fin = False
    #Comprobamos si la letra elegida está en la palabra oculta, añadimos a la lista de letras correctas si es así y activamos el contador de aciertos.
    if letra_elegida in palabra_oculta and letra_elegida not in letras_correctas:
        letras_correctas.append(letra_elegida)
        aciertos += 1
        #Imprimimos mensaje motivacional!
        print('¡Muy bien, has acertado una letra, sigue así!')
        #comprobamos que la letra elegida no fue introducida anteriormente por el judador.
    elif letra_elegida in letras_incorrectas or letra_elegida in letras_correctas:
        print("Ya introdujiste esta letra, pruebe con otra letra.")

        #Creamos el Else para imprimir la opción de letra incorrecta y activamos el contador de vidas.
    else:
        print(f"¡Vaya! La letra introducida no es correcta, tienes {vidas - 1} vidas.")
        letras_incorrectas.append(letra_elegida)
        vidas -= 1
    #Verificamos las condiciones de finalización del juego con alguna de las posibilidades y la variable fin se convierte en True.
    if vidas == 0:
        fin = perder()
    elif aciertos == letras_de_palabra:
        fin = ganar(palabra_oculta)
    #Actualización del estado del juego
    return vidas, fin, aciertos

#Función  que indica que el jugador jha perdido y muestra la palabra oculta. 
def perder():
    print("  ___")
    print(" |       |")
    print(" |       O")
    print(" |      /|\\")
    print(" |      / \\")
    print("_|_")
    print("Te has quedado sin intentos, el juego ha finalizado.")
    print("La palabra oculta era " + palabra)
    #Mensaje que indica al jugador que puede iniciar otro juego si lo desea.
    print("Si quieres jugar otra vez, dale al play.")
    return True

#Función que indica que el jugador ha ganado el juego cuando conpleta la variable palabra descubierta y la mostramos.
def ganar(palabra_descubierta):
    mostrar_nuevo_tablero(palabra_descubierta)
    print("¡¡¡Enhorabuena, has encontrado la palabra oculta!!!")

    return True

#Argumentos que nos devuelve dos valores al llamar la función elegir_palabra.
palabra, letras_de_palabra = elegir_palabra(palabras)
#Bucle que se ejecuta indicando que juego_terminado es False.
#En cada iteración se muestra el estado del tablero.
while not juego_terminado:
    print('\n' + '*' * 20 + '\n')
    mostrar_nuevo_tablero(palabra)
    print('\n')
    print('Letras incorrectas: ' + '-'.join(letras_incorrectas))
    print(f'Vidas: {intentos}')
    print('\n' + '*' * 20 + '\n')
    #Solicitamos que el jugador inserte un letra llamando la función pedir_letra.
    letra = pedir_letra()
    #Activar la función comprobar_letra y que nos devuelve los tres valores de los argumentos (intentos, terminado, aciertos).
    intentos, terminado, aciertos = comprobar_letra(letra, palabra, intentos, aciertos, letras_de_palabra)
    #Actualización de la variable juego_terminado lo la linea de codigo de arriba y cambiando el resultado para True.
    juego_terminado = terminado
```
